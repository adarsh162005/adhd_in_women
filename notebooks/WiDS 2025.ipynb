```python
def impute_numeric_with_skewness(train_Quant_new):
    # Select numeric columns
    numeric_cols = train_Quant_new.select_dtypes(include=['number']).columns

    for col in numeric_cols:
        if train_Quant_new[col].isnull().sum() > 0:  # Only process if there are missing values
            skewness = train_Quant_new[col].dropna().skew()
            if -0.5 <= skewness <= 0.5:
                # Approx. normal distribution: use mean
                fill_value = train_Quant_new[col].mean()
            else:
                # Skewed distribution: use median
                fill_value = train_Quant_new[col].median()
            
            train_Quant_new[col] = train_Quant_new[col].fillna(fill_value)

# Example usage:
impute_numeric_with_skewness(train_Quant_new)
```

## Predict Sex_F using LightGBM + Logistic Regression + Balanced Random Forest

```python
import time
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
import lightgbm as lgb

# --- Prepare target ---
y_train_sex = y_train['Sex_F']
y_val_sex = y_val['Sex_F']

# --- Feature Scaling ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_val_scaled = scaler.transform(X_val_encoded)

# --- Threshold tuner ---
def find_best_threshold(y_true, y_probs, verbose=True):
    thresholds = np.arange(0.0, 1.01, 0.01)
    best_f1 = 0
    best_thresh = 0.5
    for t in thresholds:
        preds = (y_probs >= t).astype(int)
        f1 = f1_score(y_true, preds)
        if f1 > best_f1:
            best_f1 = f1
            best_thresh = t
    if verbose:
        print(f"🔍 Best Threshold: {best_thresh:.3f}")
        print(f"🎯 Best F1 Score: {best_f1:.4f}")
    return best_thresh, best_f1

# --- Models ---
logreg_model = LogisticRegression(max_iter=10000, class_weight='balanced', random_state=42)
brf_model = BalancedRandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
lgb_model = lgb.LGBMClassifier(random_state=42)

# --- Ensemble ---
ensemble_model = VotingClassifier(
    estimators=[('logreg', logreg_model), ('brf', brf_model), ('lgb', lgb_model)],
    voting='soft'
)

# --- Training ---
start_time = time.time()
ensemble_model.fit(X_train_scaled, y_train_sex)
training_time = time.time() - start_time

# --- Predict probs & threshold ---
y_probs = ensemble_model.predict_proba(X_val_scaled)[:, 1]
best_thresh, best_f1 = find_best_threshold(y_val_sex, y_probs, verbose=False)
y_preds = (y_probs >= best_thresh).astype(int)

# --- Evaluate ---
accuracy = accuracy_score(y_val_sex, y_preds)
roc_auc = roc_auc_score(y_val_sex, y_probs)
report = classification_report(y_val_sex, y_preds)

# --- Output ---
print("\n🧬 Ensemble Evaluation for `Sex_F`:")
print(f"✅ Best Threshold  : {best_thresh:.3f}")
print(f"✅ Accuracy        : {accuracy:.4f}")
print(f"✅ F1 Score        : {best_f1:.4f}")
print(f"🔥 ROC AUC Score   : {roc_auc:.4f}")
print(f"⏱️ Training Time   : {training_time:.2f} seconds")
print("📄 Classification Report:\n", report)
```
